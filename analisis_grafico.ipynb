```python
# importamos las librerías necesarias 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## ANALISIS Y FILTRADO DE DATOS

```python
csv_file = 'CSVS STATS/todas_stats.csv'  

# Leer el archivo CSV en un DataFrame
estadisticas = pd.read_csv(csv_file)

# Pasar a data frame
df = pd.DataFrame(estadisticas)

df
```

```python
df.info()
```

```python
# Consultamos las principales estadísticas del conjunto de datos
df.describe()
```

```python
#Echamos un vistazo a la distribución de las variables de las victorias de todos los equipo mediante diagrama de cajas.

# Lista de columnas que deseas seleccionar
columnas_seleccionadas = ['victorias_esta_temp', 'victorias_ult10_temp', 'Goles Marcados Local_ult10_temp','Goles Marcados Visitante_ult10_temp']

# Seleccionar solo las columnas deseadas
df_seleccionado = df[columnas_seleccionadas]
ax = sns.boxplot(data=df_seleccionado, orient='v')

# Ajustar la orientación del texto en el eje x
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

```

Veamos las victorias de cada equipo en los ultimos 10 años

```python
# Visualizamos la distribución de los valores mediante el histograma
sns.histplot(df['victorias_ult10_temp'])
```

Observamos que 10 de los 16 equipos en octavos en la ultimas 10 temporadas no han ganado mas de 20 partidos en la Champions


```python
# Visualizamos el diagrama de caja para marcar los cuartiles y detectar los valores atípicos
sns.boxplot(data=df['victorias_ult10_temp'],  orient='v')
```

```python
# Visualizamos las relaciones de cada par de variables existentes en el conjunto de datos
sns.pairplot(data=df, height=3)
plt.show()
```

Calculo la correlacion entre las variables

```python
df_corr = df.corr()
df_corr
```

```python
plt.figure(figsize = (16,8))
sns.heatmap(df_corr, cmap='Blues', annot = True)
```

```python
# Selecciono las columnas de interés
columns_of_interest = ['victorias_esta_temp', 'victorias_ult10_temp']

# Crear un scatter plot
sns.set_theme(style="whitegrid")
sns.pairplot(df, vars=columns_of_interest, hue="equipo", palette="husl")

# Mostrar el scatter plot
plt.show()
```

```python
# Seleccionar las columnas de interés
x_column = 'porcentaje_victorias_esta_temp'
y_column = 'victorias_ult10_temp'

# Crear un scatter plot
sns.scatterplot(data=df, x=x_column, y=y_column, hue='equipo', palette='husl')

# Añadir etiquetas y título
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f'Scatter Plot entre {x_column} y {y_column}')

# Mostrar el scatter plot
plt.show()
```
