```python
#Voy a comparar las estadisticas del Real Madrid respecto al resto

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Comparacion de las victorias de esta temporada del real madrid con la media de las victorias de los demas equipos

```python
#Cargo el dataset
df = pd.read_csv('CSVS STATS/todas_stats.csv')

# Filtrar el DataFrame para obtener solo los datos del Real Madrid y del resto
real_madrid = df[df['equipo'] == 'Real Madrid']
resto = df[df['equipo'] != 'Real Madrid']

# selecciona la columna victorias_esta_temp de real_madrid y resto

victorias_real_madrid = real_madrid['victorias_esta_temp']
victorias_resto = resto['victorias_esta_temp']


# Graficar los datos
sns.histplot(victorias_real_madrid, kde=True, color='blue', label='Real Madrid', stat='density')
sns.histplot(victorias_resto, kde=True, color='red', label='Resto', stat='density')
plt.title('Distribuci√≥n de victorias')
plt.xlabel('Victorias')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

Ahora en las ultimas 10 temporadas

```python
# Filtrar el DataFrame para obtener solo los datos del Real Madrid y del resto
real_madrid = df[df['equipo'] == 'Real Madrid']
resto = df[df['equipo'] != 'Real Madrid']

# selecciona la columna victorias_esta_temp de real_madrid y resto

victorias_real_madrid = real_madrid['victorias_ult10_temp']
victorias_resto = resto['victorias_ult10_temp']


# Graficar los datos
sns.histplot(victorias_real_madrid, kde=True, color='blue', label='Real Madrid', stat='density')
sns.histplot(victorias_resto, kde=True, color='red', label='Resto', stat='density')
plt.ylim(top=0.1)  # Ajusta el valor seg√∫n sea necesario

plt.title('Distribuci√≥n de victorias')
plt.xlabel('Victorias')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

Comparacion en goles del real madrid vs la media de goles del resto de equipos

```python
# Calcular los goles totales de Real Madrid (local y visitante)
goles_real_madrid_local = real_madrid['Goles Marcados Local_esta_temp'].values[0]
goles_real_madrid_visitante = real_madrid['Goles Marcados Visitante_esta_temp'].values[0]
goles_real_madrid_totales = goles_real_madrid_local + goles_real_madrid_visitante

# Calcular la media de los goles totales del resto de equipos
media_goles_resto_local = resto['Goles Marcados Local_esta_temp'].mean()
media_goles_resto_visitante = resto['Goles Marcados Visitante_esta_temp'].mean()
media_goles_resto_totales = media_goles_resto_local + media_goles_resto_visitante

# Crear un DataFrame para las sumas de goles totales
df_goles_totales = pd.DataFrame({
    'equipo': ['Real Madrid', 'Resto de Equipos'],
    'Goles Marcados Total_esta_temp': [goles_real_madrid_totales, media_goles_resto_totales]
})

# Graficar los datos
plt.figure(figsize=(6, 4))
sns.barplot(x='equipo', y='Goles Marcados Total_esta_temp', data=df_goles_totales, palette=['blue', 'red'])
plt.title('Goles Totales como Local y Visitante esta temporada')
plt.ylabel('Goles Marcados')
plt.show()
```

```python
# Calcular los goles totales de Real Madrid en los √∫ltimos 10 a√±os (local y visitante)
goles_real_madrid_local_ult10 = real_madrid['Goles Marcados Local_ult10_temp'].values[0]
goles_real_madrid_visitante_ult10 = real_madrid['Goles Marcados Visitante_ult10_temp'].values[0]
goles_real_madrid_totales_ult10 = goles_real_madrid_local_ult10 + goles_real_madrid_visitante_ult10

# Calcular la media de los goles totales del resto de equipos en los √∫ltimos 10 a√±os
media_goles_resto_local_ult10 = resto['Goles Marcados Local_ult10_temp'].mean()
media_goles_resto_visitante_ult10 = resto['Goles Marcados Visitante_ult10_temp'].mean()
media_goles_resto_totales_ult10 = media_goles_resto_local_ult10 + media_goles_resto_visitante_ult10

# Crear un DataFrame para las sumas de goles totales en los √∫ltimos 10 a√±os
df_goles_totales_ult10 = pd.DataFrame({
    'equipo': ['Real Madrid', 'Resto de Equipos'],
    'Goles Marcados Total_ult10_temp': [goles_real_madrid_totales_ult10, media_goles_resto_totales_ult10]
})

# Graficar los datos
plt.figure(figsize=(8, 5))
sns.barplot(x='equipo', y='Goles Marcados Total_ult10_temp', data=df_goles_totales_ult10, palette=['blue', 'red'])
plt.title('Goles Totales como Local y Visitante en los √∫ltimos 10 a√±os')
plt.ylabel('Goles Marcados')
plt.show()
```

Comparacion de titulos UCL ganados el Real Madrid vs la suma de los titulos de todos los clasificados a octavos

```python
# Seleccionar la columna 'titulos_UCL_ganados' de real_madrid y resto
titulos_real_madrid = real_madrid['titulos_UCL_ganados'].values[0]
media_titulos_resto = resto['titulos_UCL_ganados'].sum()

# Crear un DataFrame para la media del resto
df_media_resto = pd.DataFrame({'equipo': ['Resto de Equipos'], 'titulos_UCL_ganados': [media_titulos_resto]})

# Concatenar el DataFrame de Real Madrid con el de la media del resto
df_plot = pd.concat([real_madrid, df_media_resto])

# Graficar los datos
plt.figure(figsize=(10, 6))
sns.barplot(x='equipo', y='titulos_UCL_ganados', data=df_plot, palette=['blue', 'red'])
plt.title('Comparaci√≥n de T√≠tulos de la UCL')
plt.ylabel('N√∫mero de T√≠tulos UCL')
plt.show()
```
